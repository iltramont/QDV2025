```python
import pandas as pd
```

```python
# Load
data = pd.read_csv("20251201_QDV2025_001.csv")
```

```python
data.head()
```

```python
data.rename(columns={'DENOMINAZIONE CORRENTE': 'PROVINCIA'}, inplace=True)
```

```python
print(f"Numero indicatori presi in esame: {len(data['INDICATORE'].unique())}")
print(f"Numero province esaminate: {len(data['PROVINCIA'].unique())}")
```

```python
# Verifica univocità
for c in data.columns[:4]:
    print(c, len(data[c].unique()))
```

```python
tabella_province = data.iloc[:, :4]
tabella_province = tabella_province.drop_duplicates().reset_index(drop=True)
tabella_province.columns = ['NOME_ISTAT', 'COD_NUTS', 'COD_ISTAT', 'PROVINCIA']
display(tabella_province)
```

```python
tabella_indicatori = data.iloc[:, 5:]
tabella_indicatori = tabella_indicatori.drop_duplicates().reset_index(drop=True)
tabella_indicatori.columns = ['INDICATORE', 'UNITA_MISURA', 'ANNO_RIFERIMENTO', 'FONTE']
display(tabella_indicatori)
```

```python
tabella_punteggi = data[['PROVINCIA', 'INDICATORE', 'VALORE']]
tabella_punteggi = tabella_punteggi.pivot(index='PROVINCIA', columns='INDICATORE', values='VALORE')
display(tabella_punteggi)
```
